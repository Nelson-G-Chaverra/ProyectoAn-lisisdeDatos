# CREACIÓN DE NUEVAS VARIABLES

```python
# NUEVA VARIABLE PARA LA PERSONA QUE VIAJABA "SOLO"
train['FlagSolo'] = np.where(
    (train['SibSp'] == 0) & (train['Parch'] == 0), 1, 0)
```

```python
train[['Survived', 'Sex', 'Age', 'Pclass', 'FlagSolo']].head(3)
```

```python
# VARIABLE DEPENDIENTE
Y_train = train['Survived']

# PREPROCESAMIENTO DE VARIABLES INDEPENDIENTES
features = ['Sex', 'Age', 'Pclass', 'FlagSolo']
X_train = train[features]

print(Y_train.shape, X_train.shape)
```

# MODELO DE ANÁLISIS REGRESIÓN LOGISTICA Y ARBOL DE DECISIÓN

```python
# MODELO REGRESION LOGISTICA
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
```

```python
# MODELO ARBOL DE DECISIÓN
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def conf_mat_acc(modelo, X_train, Y_train):
    # Generar la matriz de confusión
    cm = confusion_matrix(Y_train, modelo.predict(X_train))
    
    # Calcular la precisión
    true_pred = cm[0, 0] + cm[1, 1]
    total_data = np.sum(cm)
    accuracy = true_pred / total_data
    print('accuracy: ', np.round(accuracy, 2))
    
    # Graficar la matriz de confusión
    sns.heatmap(cm, annot=True, fmt='d', cmap=plt.cm.Blues, xticklabels=[0, 1], yticklabels=[0, 1])
    plt.xlabel('Predicción')
    plt.ylabel('Verdadero')
    plt.show()
```

```python
conf_mat_acc(logreg, X_train, Y_train)
```

```python
conf_mat_acc(decision_tree, X_train, Y_train)
```
